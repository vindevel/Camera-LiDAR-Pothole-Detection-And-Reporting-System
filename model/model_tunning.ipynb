```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from glob import glob
from collections import Counter

# 분석할 YOLO 형식 라벨 경로 (dataset1, 3, 4 경로만 변경해주면 됨)
label_dir = "/content/dataset/train/labels"

# 라벨 파싱 함수 (변환 없음)
def parse_yolo_labels(label_path):
    boxes = []
    with open(label_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) == 5:
                class_id, x_center, y_center, width, height = map(float, parts)
                boxes.append([int(class_id), x_center, y_center, width, height])
    return boxes

# 라벨 처리
label_files = glob(os.path.join(label_dir, "*.txt"))
all_boxes, bbox_count_per_image, class_ids = [], [], []

for file in label_files:
    boxes = parse_yolo_labels(file)
    if boxes:
        bbox_count_per_image.append(len(boxes))
        for box in boxes:
            all_boxes.append(box)
            class_ids.append(box[0])

# numpy 변환 및 검증
box_array = np.array(all_boxes)
if box_array.ndim == 2 and box_array.shape[1] == 5:
    x_centers = box_array[:, 1]
    y_centers = box_array[:, 2]
    widths = box_array[:, 3]
    heights = box_array[:, 4]
    aspect_ratios = widths / heights

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    sns.histplot(x=x_centers, y=y_centers, bins=50, pmax=0.9, ax=axes[0, 0], cbar=True)
    axes[0, 0].set_title("▶ Bounding Box Center Distribution")
    axes[0, 0].set_xlabel("x_center (normalized)")
    axes[0, 0].set_ylabel("y_center (normalized)")

    axes[0, 1].hist(widths, bins=40, color='skyblue')
    axes[0, 1].set_title("▶ Width Distribution")
    axes[0, 1].set_xlabel("width")
    axes[0, 1].set_ylabel("count")

    axes[0, 2].hist(heights, bins=40, color='salmon')
    axes[0, 2].set_title("▶ Height Distribution")
    axes[0, 2].set_xlabel("height")
    axes[0, 2].set_ylabel("count")

    axes[1, 0].hist(bbox_count_per_image, bins=range(1, max(bbox_count_per_image) + 2), color='gray', rwidth=0.8)
    axes[1, 0].set_title("▶ Bounding Box Count per Image")
    axes[1, 0].set_xlabel("# of boxes")
    axes[1, 0].set_ylabel("image count")

    class_counter = Counter(class_ids)
    axes[1, 1].bar(class_counter.keys(), class_counter.values(), color='purple')
    axes[1, 1].set_title("▶ Class Distribution")
    axes[1, 1].set_xlabel("Class ID")
    axes[1, 1].set_ylabel("Count")

    axes[1, 2].hist(aspect_ratios, bins=40, color='orange')
    axes[1, 2].set_title("▶ Aspect Ratio (width / height)")
    axes[1, 2].set_xlabel("Aspect Ratio")
    axes[1, 2].set_ylabel("count")

    plt.tight_layout()
    plt.show()
else:
    print(f"box_array가 비어 있거나 예상한 5열 구조가 아닙니다. 현재 shape: {box_array.shape}")
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from glob import glob
from collections import Counter

label_dir = "/content/dataset2/train/labels"

# 데이터셋 2의 경우 bbox가 아닌 polygon이라 따로 변환 필요
# polygon → bbox 변환 함수
def polygon_to_yolo_bbox(parts):
    try:
        coords = list(map(float, parts[1:]))  # 좌표들만 가져옴
        x_coords = coords[::2]
        y_coords = coords[1::2]
        if not x_coords or not y_coords:
            return None
        x_min, x_max = min(x_coords), max(x_coords)
        y_min, y_max = min(y_coords), max(y_coords)
        x_center = (x_min + x_max) / 2
        y_center = (y_min + y_max) / 2
        width = x_max - x_min
        height = y_max - y_min
        return [int(float(parts[0])), x_center, y_center, width, height]
    except Exception as e:
        print(f"변환 실패: {parts} → {e}")
        return None

# 라벨 파싱
def parse_polygon_labels(label_path):
    boxes = []
    with open(label_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 9 and (len(parts) - 1) % 2 == 0:
                box = polygon_to_yolo_bbox(parts)
                if box:
                    boxes.append(box)
    return boxes

# 라벨 처리
label_files = glob(os.path.join(label_dir, "*.txt"))
all_boxes, bbox_count_per_image, class_ids = [], [], []

for file in label_files:
    boxes = parse_polygon_labels(file)
    if boxes:
        bbox_count_per_image.append(len(boxes))
        for box in boxes:
            all_boxes.append(box)
            class_ids.append(box[0])

# numpy 변환 및 검증
box_array = np.array(all_boxes)
if box_array.ndim == 2 and box_array.shape[1] == 5:
    x_centers = box_array[:, 1]
    y_centers = box_array[:, 2]
    widths = box_array[:, 3]
    heights = box_array[:, 4]
    aspect_ratios = widths / heights

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    sns.histplot(x=x_centers, y=y_centers, bins=50, pmax=0.9, ax=axes[0, 0], cbar=True)
    axes[0, 0].set_title("▶ Bounding Box Center Distribution")
    axes[0, 0].set_xlabel("x_center (normalized)")
    axes[0, 0].set_ylabel("y_center (normalized)")

    axes[0, 1].hist(widths, bins=40, color='skyblue')
    axes[0, 1].set_title("▶ Width Distribution")
    axes[0, 1].set_xlabel("width")
    axes[0, 1].set_ylabel("count")

    axes[0, 2].hist(heights, bins=40, color='salmon')
    axes[0, 2].set_title("▶ Height Distribution")
    axes[0, 2].set_xlabel("height")
    axes[0, 2].set_ylabel("count")

    axes[1, 0].hist(bbox_count_per_image, bins=range(1, max(bbox_count_per_image) + 2), color='gray', rwidth=0.8)
    axes[1, 0].set_title("▶ Bounding Box Count per Image")
    axes[1, 0].set_xlabel("# of boxes")
    axes[1, 0].set_ylabel("image count")

    class_counter = Counter(class_ids)
    axes[1, 1].bar(class_counter.keys(), class_counter.values(), color='purple')
    axes[1, 1].set_title("▶ Class Distribution")
    axes[1, 1].set_xlabel("Class ID")
    axes[1, 1].set_ylabel("Count")

    axes[1, 2].hist(aspect_ratios, bins=40, color='orange')
    axes[1, 2].set_title("▶ Aspect Ratio (width / height)")
    axes[1, 2].set_xlabel("Aspect Ratio")
    axes[1, 2].set_ylabel("count")

    plt.tight_layout()
    plt.show()
else:
    print(f"❗ box_array가 비어 있거나 예상한 5열 구조가 아닙니다. 현재 shape: {box_array.shape}")
```

## 이상치 필터 기준
* MAX_BBOX_COUNT = 5      # 한 이미지당 최대 박스 수 (5개 미만)

## dataset1의 이상치 제거

```python
# dataset1 train 이상치 제거
import os
from glob import glob
from shutil import copy2

# 경로 설정 (필요 시 수정)
label_dir = "/content/Enes_Guvercin_Bozuk_Yol_Tespiti-1/train/labels"
image_dir = "/content/Enes_Guvercin_Bozuk_Yol_Tespiti-1/train/images"
output_label_dir = "/content/dataset1/train/labels"
output_image_dir = "/content/dataset1/train/images"

# 필터 기준
MAX_BBOX_COUNT = 5      # 한 이미지당 최대 박스 수 (5개 미만)
MAX_BBOX_AREA = 0.8     # 박스 면적 기준 (w * h < 0.8)
MAX_BBOX_WIDTH = 0.8    # 박스 너비 기준 (w < 0.8)
MAX_BBOX_HEIGHT = 0.8   # 박스 높이 기준 (h < 0.8)

# 출력 폴더 생성
os.makedirs(output_label_dir, exist_ok=True)
os.makedirs(output_image_dir, exist_ok=True)

# 통계 초기화
total = 0
kept = 0
removed = 0

# 전체 라벨 파일 순회
label_files = sorted(glob(os.path.join(label_dir, "*.txt")))

for label_file in label_files:
    with open(label_file, "r") as f:
        lines = f.readlines()

    total += 1
    valid = True

    # 조건 1: 박스 개수 제한
    if len(lines) >= MAX_BBOX_COUNT:
        valid = False
    else:
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
            _, x, y, w, h = map(float, parts)
            area = w * h

            # 조건 2: 면적 또는 개별 크기 초과 시 제거
            if area > MAX_BBOX_AREA or w > MAX_BBOX_WIDTH or h > MAX_BBOX_HEIGHT:
                valid = False
                break

    # 유지 여부 결정
    if valid:
        # 파일 복사
        fname = os.path.basename(label_file)
        copy2(label_file, os.path.join(output_label_dir, fname))

        img_path_jpg = os.path.join(image_dir, fname.replace(".txt", ".jpg"))
        img_path_png = os.path.join(image_dir, fname.replace(".txt", ".png"))
        img_path = img_path_jpg if os.path.exists(img_path_jpg) else img_path_png

        if os.path.exists(img_path):
            copy2(img_path, os.path.join(output_image_dir, os.path.basename(img_path)))
        kept += 1
    else:
        removed += 1

# 결과 요약 출력
print(f"총 파일 수: {total}")
print(f"유지된 파일 수: {kept}")
print(f"제거된 파일 수: {removed}")
```

```python
# dataset2 test 이상치 제거

# 경로 설정 (필요 시 수정)
label_dir = "/content/Pothole-Detection-Project-3/test/labels"
image_dir = "/content/Pothole-Detection-Project-3/test/images"
output_label_dir = "/content/dataset2/test/labels"
output_image_dir = "/content/dataset2/test/images"

# 필터 기준
MAX_BBOX_COUNT = 5
MAX_BBOX_AREA = 0.8
MAX_BBOX_WIDTH = 0.8
MAX_BBOX_HEIGHT = 0.8

# 출력 폴더 생성
os.makedirs(output_label_dir, exist_ok=True)
os.makedirs(output_image_dir, exist_ok=True)

# Polygon → BBox 변환 함수
def polygon_to_bbox(parts):
    try:
        coords = list(map(float, parts[1:]))
        x_coords = coords[::2]
        y_coords = coords[1::2]
        if not x_coords or not y_coords:
            return None
        x_min, x_max = min(x_coords), max(x_coords)
        y_min, y_max = min(y_coords), max(y_coords)
        x_center = (x_min + x_max) / 2
        y_center = (y_min + y_max) / 2
        width = x_max - x_min
        height = y_max - y_min
        return [int(float(parts[0])), x_center, y_center, width, height]
    except:
        return None

# 통계 초기화
total = 0
kept = 0
removed = 0

# 전체 라벨 파일 순회
label_files = sorted(glob(os.path.join(label_dir, "*.txt")))

for label_file in label_files:
    with open(label_file, "r") as f:
        lines = f.readlines()

    total += 1
    valid = True
    converted_boxes = []

    for line in lines:
        parts = line.strip().split()
        if len(parts) >= 9 and (len(parts) - 1) % 2 == 0:
            box = polygon_to_bbox(parts)
            if box:
                _, x_center, y_center, w, h = box
                area = w * h
                # 면적, 너비, 높이 체크
                if area >= MAX_BBOX_AREA:
                    valid = False
                    break
                converted_boxes.append(box)

    # 박스 개수 체크
    if len(converted_boxes) == 0 or len(converted_boxes) >= MAX_BBOX_COUNT:
        valid = False

    # 유지 여부 결정
    if valid:
        # 복사
        fname = os.path.basename(label_file)
        copy2(label_file, os.path.join(output_label_dir, fname))

        img_path_jpg = os.path.join(image_dir, fname.replace(".txt", ".jpg"))
        img_path_png = os.path.join(image_dir, fname.replace(".txt", ".png"))
        img_path = img_path_jpg if os.path.exists(img_path_jpg) else img_path_png

        if os.path.exists(img_path):
            copy2(img_path, os.path.join(output_image_dir, os.path.basename(img_path)))
        kept += 1
    else:
        removed += 1

# 결과 출력
print(f"총 파일 수: {total}")
print(f"유지된 파일 수: {kept}")
print(f"제거된 파일 수: {removed}")
```

## 구성1

```python
import os
import random
from glob import glob
from shutil import copy2
from collections import defaultdict

# 사용할 데이터셋 경로
base_paths = {
    "dataset2": "/content/dataset2",
    "dataset3": "/content/dataset3",
    "dataset4": "/content/dataset4"
}

# Dataset 간 비중
dataset_ratios = {
    "dataset3": 0.6,
    "dataset4": 0.1,
    "dataset2": 0.3
}

# Split 비율
split_ratios = {
    "train": 0.8,
    "val": 0.1,
    "test": 0.1
}

output_base = "./Dataset_Merged"
for split in ["train", "val", "test"]:
    for sub in ["images", "labels"]:
        os.makedirs(os.path.join(output_base, split, sub), exist_ok=True)

# 단일 bbox 수 세기 함수
def count_bboxes(label_path):
    with open(label_path, "r") as f:
        return len(f.readlines())

# bbox 기반 샘플링
def sample_by_bbox(label_files, target_bbox_count):
    random.seed(42)
    random.shuffle(label_files)
    selected = []
    bbox_sum = 0
    for path in label_files:
        count = count_bboxes(path)
        if bbox_sum + count > target_bbox_count:
            break
        selected.append((path, count))
        bbox_sum += count
    return selected, bbox_sum

# 복사 함수
def copy_pair(label_path, src_img_dir, dst_label_dir, dst_img_dir):
    fname = os.path.basename(label_path)
    img_path = os.path.join(src_img_dir, fname.replace(".txt", ".jpg"))
    if os.path.exists(img_path):
        copy2(label_path, os.path.join(dst_label_dir, fname))
        copy2(img_path, os.path.join(dst_img_dir, os.path.basename(img_path)))

# 전체 bbox 기준 계산
total_bbox = 0
bbox_files_by_dataset = {}
for name, base in base_paths.items():
    label_files = glob(os.path.join(base, "train/labels", "*.txt")) + \
                  glob(os.path.join(base, "valid/labels", "*.txt"))
    bbox_counts = [(f, count_bboxes(f)) for f in label_files]
    bbox_files_by_dataset[name] = bbox_counts
    total_bbox += sum(b for _, b in bbox_counts)

# 병합 및 분할
summary = defaultdict(int)
selection_log = {}

for name, files in bbox_files_by_dataset.items():
    dataset_target = int(total_bbox * dataset_ratios[name])
    selected_tuples, _ = sample_by_bbox([f for f, _ in files], dataset_target)

    total_files = len(selected_tuples)
    split_train = int(total_files * split_ratios["train"])
    split_val = int(total_files * split_ratios["val"])

    split_sets = {
        "train": selected_tuples[:split_train],
        "val": selected_tuples[split_train:split_train + split_val],
        "test": selected_tuples[split_train + split_val:]
    }

    selection_log[name] = {}
    for split in ["train", "val", "test"]:
        bbox_sum = 0
        for f, count in split_sets[split]:
            bbox_sum += count
            src_img_dir = os.path.dirname(f).replace("labels", "images")
            copy_pair(f, src_img_dir,
                      os.path.join(output_base, split, "labels"),
                      os.path.join(output_base, split, "images"))
            summary[split] += 1
        selection_log[name][f"{split}_files"] = len(split_sets[split])
        selection_log[name][f"{split}_bboxes"] = bbox_sum

# 요약 출력
print("병합 및 분할 완료 (비율: Dataset2:60%, 3:10%, 4:30% → Train:Val:Test = 80:10:10)")
for split in ["train", "val", "test"]:
    print(f"✅ {split.upper()} 라벨 수: {summary[split]}")
print("\n📊 데이터셋별 상세:")
for name, stats in selection_log.items():
    print(f"\n📁 {name.upper()}")
    for k, v in stats.items():
        print(f"  - {k}: {v}")
```

## 구성 2

```python
import os
import random
from glob import glob
from shutil import copy2

RANDOM_SEED = 42
random.seed(RANDOM_SEED)

datasets = {
    "dataset1": "/content/dataset1",
    "dataset2": "/content/dataset2",
    "dataset3": "/content/dataset3",
    "dataset4": "/content/dataset4",
}

sampling_targets = {
    "dataset1": {"train": 3000, "valid": 376, "test": 376},
    "dataset2": {"train": 8200, "valid": 1028, "test": 1028},
    "dataset3": {"train": 649, "valid": 75, "test": 75},
    "dataset4": {"train": 8200, "valid": 1028, "test": 1028},
}

output_base = "/content/Dataset"

for split in ["train", "valid", "test"]:
    for sub in ["images", "labels"]:
        os.makedirs(os.path.join(output_base, split, sub), exist_ok=True)

def count_objects(label_path):
    with open(label_path, "r") as f:
        return len(f.readlines())

def sample_by_object_count(label_files, target_count):
    random.shuffle(label_files)
    selected = []
    object_sum = 0
    for label_path in label_files:
        count = count_objects(label_path)
        if object_sum + count > target_count:
            break
        selected.append(label_path)
        object_sum += count
    return selected, object_sum

result_summary = {"train": {"files": 0, "objects": 0},
                  "valid": {"files": 0, "objects": 0},
                  "test": {"files": 0, "objects": 0}}

for dataset_name, dataset_path in datasets.items():
    for split in ["train", "valid", "test"]:
        label_dir = os.path.join(dataset_path, split, "labels")
        image_dir = os.path.join(dataset_path, split, "images")
        label_files = sorted(glob(os.path.join(label_dir, "*.txt")))

        if not label_files:
            print(f"{dataset_name} - {split} 경로에 라벨 파일이 없습니다: {label_dir}")
            continue

        target_object_count = sampling_targets[dataset_name][split]
        selected_files, total_objects = sample_by_object_count(label_files, target_object_count)

        print(f"{dataset_name} - {split}: {len(selected_files)} files selected, {total_objects} objects")

        result_summary[split]["files"] += len(selected_files)
        result_summary[split]["objects"] += total_objects

        for label_file in selected_files:
            fname = os.path.basename(label_file)
            copy2(label_file, os.path.join(output_base, split, "labels", fname))

            img_jpg = os.path.join(image_dir, fname.replace(".txt", ".jpg"))
            img_png = os.path.join(image_dir, fname.replace(".txt", ".png"))
            img_path = img_jpg if os.path.exists(img_jpg) else img_png

            if os.path.exists(img_path):
                copy2(img_path, os.path.join(output_base, split, "images", os.path.basename(img_path)))

print("\n▶ 분할 요약")
for split in ["train", "valid", "test"]:
    print(f"{split.capitalize()} - 파일 수: {result_summary[split]['files']}, 바운딩 박스 수: {result_summary[split]['objects']}")
```

```python
import os
import random
from glob import glob
from shutil import copy2

RANDOM_SEED = 42
random.seed(RANDOM_SEED)

datasets = {
    "dataset1": "/content/dataset1",
    "dataset2": "/content/dataset2",
    "dataset3": "/content/dataset3",
    "dataset4": "/content/dataset4",
}

sampling_targets = {
    "dataset1": {"train": 2400, "valid": 300, "test": 300},
    "dataset2": {"train": 8200, "valid": 1026, "test": 1026},
    "dataset3": {"train": 649, "valid": 75, "test": 75},
    "dataset4": {"train": 4751, "valid": 592, "test": 592},
}

output_base = "/content/final_Dataset"

for split in ["train", "valid", "test"]:
    for sub in ["images", "labels"]:
        os.makedirs(os.path.join(output_base, split, sub), exist_ok=True)

def count_objects(label_path):
    with open(label_path, "r") as f:
        return len(f.readlines())

def sample_by_object_count(label_files, target_count):
    random.shuffle(label_files)
    selected = []
    object_sum = 0
    for label_path in label_files:
        count = count_objects(label_path)
        if object_sum + count > target_count:
            break
        selected.append(label_path)
        object_sum += count
    return selected, object_sum

result_summary = {"train": {"files": 0, "objects": 0},
                  "valid": {"files": 0, "objects": 0},
                  "test": {"files": 0, "objects": 0}}

for dataset_name, dataset_path in datasets.items():
    for split in ["train", "valid", "test"]:
        label_dir = os.path.join(dataset_path, split, "labels")
        image_dir = os.path.join(dataset_path, split, "images")
        label_files = sorted(glob(os.path.join(label_dir, "*.txt")))

        if not label_files:
            print(f"{dataset_name} - {split} 경로에 라벨 파일이 없습니다: {label_dir}")
            continue

        target_object_count = sampling_targets[dataset_name][split]
        selected_files, total_objects = sample_by_object_count(label_files, target_object_count)

        print(f"{dataset_name} - {split}: {len(selected_files)} files selected, {total_objects} objects")

        result_summary[split]["files"] += len(selected_files)
        result_summary[split]["objects"] += total_objects

        for label_file in selected_files:
            fname = os.path.basename(label_file)
            copy2(label_file, os.path.join(output_base, split, "labels", fname))

            img_jpg = os.path.join(image_dir, fname.replace(".txt", ".jpg"))
            img_png = os.path.join(image_dir, fname.replace(".txt", ".png"))
            img_path = img_jpg if os.path.exists(img_jpg) else img_png

            if os.path.exists(img_path):
                copy2(img_path, os.path.join(output_base, split, "images", os.path.basename(img_path)))

print("\n▶ 분할 요약")
for split in ["train", "valid", "test"]:
    print(f"{split.capitalize()} - 파일 수: {result_summary[split]['files']}, 바운딩 박스 수: {result_summary[split]['objects']}")
```

```python
import time

def load_labels(label_file):
    boxes = []
    with open(label_file, 'r') as f:
        for line in f:
            cls, x, y, w, h = map(float, line.strip().split())
            boxes.append([x, y, w, h])
    return boxes

def run_tensorrt_inference(session, input_data):
    input_name = session.get_inputs()[0].name
    start_time = time.time()
    outputs = session.run(None, {input_name: input_data})
    duration = time.time() - start_time
    return outputs, duration

def compute_iou(box1, box2):
    # box: [xc, yc, w, h]
    x1_min = box1[0] - box1[2] / 2
    x1_max = box1[0] + box1[2] / 2
    y1_min = box1[1] - box1[3] / 2
    y1_max = box1[1] + box1[3] / 2

    x2_min = box2[0] - box2[2] / 2
    x2_max = box2[0] + box2[2] / 2
    y2_min = box2[1] - box2[3] / 2
    y2_max = box2[1] + box2[3] / 2

    xi1 = max(x1_min, x2_min)
    yi1 = max(y1_min, y2_min)
    xi2 = min(x1_max, x2_max)
    yi2 = min(y1_max, y2_max)

    inter_area = max(xi2 - xi1, 0) * max(yi2 - yi1, 0)
    box1_area = (x1_max - x1_min) * (y1_max - y1_min)
    box2_area = (x2_max - x2_min) * (y2_max - y2_min)
    union_area = box1_area + box2_area - inter_area

    return inter_area / union_area if union_area > 0 else 0
```
